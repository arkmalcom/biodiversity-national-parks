## Analysis

### Visualizing the species information data

Let's begin by showing a breakdown of the different conservation statuses and how many total entries there are for each.

```python
species_info.conservation_status.value_counts()
```

We can see that most species are considered `Least Concern` with a total of 5633. The least represented category is `In Recovery` with a total of 4.

Next, let's create a stacked bar graph that will show not just a count of the different conservation statuses, but divide them by category. It's important to not include species of `Least Concern` as there is a significantly larger proportion of those compared to anything else, which will make it the only thing visible in the bar graph.

```python
conservation_category = species_info[species_info.conservation_status != 'Least Concern']\
                        .groupby(['conservation_status', 'category']).scientific_name.count().unstack()
sns.set_style("darkgrid")
sns.set_context("talk")
sns.set_palette(sns.color_palette("Set1"))
ax = conservation_category.plot(kind='bar', stacked=True, figsize=(14,8))
ax.set_xlabel('Conservation Status')
ax.set_ylabel('Species Count')
ax.set_title("Species Count by Conservation Status")
```

From this we can see that `Mammal` is the most endangered category, very closely followed by `Bird`. There are no values for `Nonvascular Plant` or `Reptile` in the endangered status. Is it possible that these species are less likely to become endangered than others? To test for this, we'll need a percentage we can use and put into a contingency table. The first step is create a column that tracks if a species is considered protected or not. Any category other than `Least Concern` will count as a protected species.

```python
species_info['protected'] = species_info.conservation_status != 'Least Concern'
species_info.groupby(['category', 'protected']).scientific_name.count()
```

This table can be difficult to read, not to mention raw counts are difficult to visualize, especially when some values are much larger than others. As we planned to use percentages in our contingency table, we will use a `lambda` function that takes the values in each row, divides them by the sum of the values in said row, and then multiplies by 100. This will leave us with a value between 0 and 100, and the values of `True` and `False` should add up to 100.

```python
species_crosstab = pd.crosstab(species_info.category, species_info.protected).apply(lambda row: row/row.sum() * 100, axis=1)
species_crosstab
```

We can now use a Chi-Squared test to see if certain species are more likely to be endangered in comparison to others. We could do a general test that includes this entire contingency table.

```python
from scipy.stats import chi2_contingency

_, pval, _, _ = chi2_contingency(species_crosstab)
pval
```

As `chi2_contingency` returns four values, and we're only really interested in the p value, we can assign all the other ones to blank variables. The resulting p value from this test is **0.00002577**, which is well below **0.05**, indicating a significant relationship between species and endangered status. This would indicate that some species require more protection than others.

However, it's also possible to make comparisons in pairs. For instance, we can ask ourselves: Do amphibians require more protection than birds? We can use a for loop to go through some(though not all) comparisons quickly.

```python
for i,j in zip(range(0,7), range(2,7)):
    comparison = species_crosstab[i:j]
    _, pval, _, _ = chi2_contingency(comparison)
    print("The p value for the comparison between {0} and {1} is: {2}".format(comparison.index[0], comparison.index[1], pval))
```

Though this test doesn't test every possible combination, we can once again see that certain species require more protection than others. For instance, we see that the difference between amphibians and birds is not significant and thus we can assume that in thhis case, one doesn't require more protection than the other. However, in the case of Mammals and Nonvascular plants, we observe a significant p value, indicative of the fact that Mammals require more protection than Nonvascular plants.

### Visualizing the observations data

We can start with something simple, such as simply seeing which park has the most observations. 

```python
sns.set_style("darkgrid")
sns.set_context("talk")
plt.figure(figsize=(14,8))
ax2 = sns.barplot(data=observations, x='park_name', y='observations', estimator=sum,\
                  ci=None)
ax2.set_xlabel('Park Name')
ax2.set_ylabel('Observations (in millions of sightings)')
ax2.set_xticklabels(ax2.get_xticklabels(), rotation=30)
ax2.set_title("Observations of Species at the Different National Parks")
```

By looking at the bar graph, we can see that **Yellowstone National Park** has the most amount of observations, followed by **Yosemite National Park**, with **Great Smoky Mountains National Park** containing the least amount of observations.

Next we can see if one park is more likely to observe endangered species than another. As we have already created a `protected` column for our `species_info` dataframe, we can simply merge it with our `observations` dataframe and count observations of protected species.

```python
merged_data = observations.merge(species_info)
merged_data.head()
```

Now we have a table that not only contains the `protected` column, but also an `observations` column, making it very simple to count how many protected species were observed at each park.

```python
sns.set_style("darkgrid")
sns.set_context("talk")
plt.figure(figsize=(14,8))
ax3 = sns.barplot(data=merged_data, x='park_name', y='observations', hue='protected', estimator=sum, ci=None)
ax3.set_xlabel("Park Name")
ax3.set_xticklabels(ax3.get_xticklabels(), rotation=30)
ax3.set_yscale('log')
ax3.set_ylabel("Observations (log scale)")
ax3.set_title("Observations of Protected vs. Nonprotected Species by Park")
```

We once again see **Yellowstone National Park** take the lead, though this time, the difference is much smaller. As far as protected species go, **Yosemite National Park** and **Yellowstone National Park** come fairly close in regards to observations. A log scale was used for the Y-axis as the pure numerical difference between values was significant in some cases.

If you want to have a higher chance of observing a protected species, **Yellowstone National Park** is the place to go.

### Further analysis

This dataset contains a lot of useful data. For example, using the `common_names` column, we could see where we're more likely to observe a certain species (such as a Mink, or Aurochs). We'll need to use RegEx in order to test this. As we already have a dataframe that contains both observation data and species information, we can simply add a new column to it to check if a row contains information on a species we're interested in. Let's say bisons are our favorite type of animal, so we want to go somewhere where we're most likely to see one. Let's add a column that uses RegEx to check if `common_names` contains **Bison**.

```python
merged_data['is_bison'] = merged_data.common_names.str.contains(r"\bBison\b", regex=True)
bison_observations = merged_data[merged_data.is_bison]
bison_observations
```

This table shows observations where `is_bison` is `True`. From here, we can once again simply add up the amount of observations and see where it's most likely we'll see any type of bison. 

```python
sns.set_style("darkgrid")
sns.set_context("talk")
sns.set_palette(sns.color_palette("OrRd"))
plt.figure(figsize=(14,8))
ax4 = sns.barplot(data=bison_observations, x='park_name', y='observations', estimator=sum, \
                  ci=None, order=bison_observations.sort_values('observations').park_name.unique())
ax4.set_xticklabels(ax4.get_xticklabels(), rotation=30)
ax4.set_xlabel("Park Name")
ax4.set_ylabel("Observations")
ax4.set_title("Bison Observations by Park")
```

It seems that if we want to have the highest probability of being able to see any type of bison observed in the dataset, we will have to go to **Yellowstone National Park**
